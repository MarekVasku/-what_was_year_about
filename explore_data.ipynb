```python
# Import required libraries
import os
import sys
import pandas as pd
from pathlib import Path

# Add the src directory to Python path to import data_utils
src_path = Path.cwd() / 'src'
if src_path not in sys.path:
    sys.path.append(str(src_path))

try:
    import src.data_utils
    print("Successfully imported data_utils")
except ImportError as e:
    print(f"Failed to import data_utils: {e}")
    print(f"Python path: {sys.path}")
```

```python
# Check if we can access the data
try:
    result = data_utils.get_data_cached()
    print("Successfully loaded data")
    
    # Just work with the first item if it's a DataFrame
    df = result[0] if isinstance(result[0], pd.DataFrame) else result
    
    print("\nDataFrame Info:")
    print(df.info())
    print("\nColumns:")
    print(df.columns.tolist())
except Exception as e:
    print(f"Error loading data: {e}")
```
